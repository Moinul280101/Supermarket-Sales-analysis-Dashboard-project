```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("supermarket_sales -.csv")
df.head()
```

```python
# basis information about data
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
#check null data
pd.isna(df).sum()
```

```python
#Total sales by the branch
sns.barplot(x=df['Branch'],y=df['Total'])
plt.title("Total sales by the branches")
```

```python
#Total product line sales
sns.boxplot(y=df['Product line'],x=df['Total'])
plt.title("Product line sales")
```

```python
#payment method sistribution
sns.countplot(x=df['Payment'])
plt.title("Payment Metod Distribution")
```

```python
#Total sales by the citys
sns.barplot(y=df['City'],x=df['Total'],estimator=sum)
plt.title("total sales by the city")
```

```python

sns.barplot(x=df['Rating'].mean(),y=df['Gender'],hue=df['Gender'])
plt.title("avarage rating by the gender")
```

```python
# gross income by the product
sns.barplot(x=df['gross income'],y=df['Product line'],hue=df['Product line'])
plt.title(" gross income by the product")
```

```python
#feature extraction
df['Date']=pd.to_datetime(df['Date'])
           
```

```python
df['month']=df['Date'].dt.month_name()
```

```python
#monthly trend
sns.lineplot(x=df['month'],y=df['Total'])
```

```python
#Save cleaned data
df.to_csv("cleaned_supermarket_sales.csv")
```
